```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

NO_CORNERS = 78  # Target number of corners to match

def first_correct_winsize(I):
    print(f"Task 1: Detecting corners for image of shape: {I.shape}")
    I_blur = cv2.GaussianBlur(I, (5, 5), 0)
    gray = cv2.cvtColor(I_blur, cv2.COLOR_BGR2GRAY)
    gray = np.float32(gray)

    best_win_size = -1
    best_corner_count = float('inf')
    best_params = {}

    for win_size_ in [2, 4, 6, 8, 10, 12]:
        for ksize_ in [3, 5, 7]:
            for alpha_ in [0.04, 0.06, 0.08]:
                dst = cv2.cornerHarris(gray, win_size_, ksize_, alpha_)
                dst = cv2.dilate(dst, None)
                thresh = 0.02 * dst.max()
                dst_thresh = (dst > thresh).astype(np.uint8) * 255

                num_labels, _, _, _ = cv2.connectedComponentsWithStats(dst_thresh)
                nc = num_labels - 1

                print(f"  WinSize: {win_size_}, kSize: {ksize_}, Alpha: {alpha_}, Corners: {nc}")

                if nc == NO_CORNERS:
                    print("  Found exact match!")
                    return win_size_
                elif abs(nc - NO_CORNERS) < abs(best_corner_count - NO_CORNERS):
                    best_corner_count = nc
                    best_win_size = win_size_
                    best_params = {"win_size": win_size_, "ksize": ksize_, "alpha": alpha_}

    if best_win_size != -1:
        print(f"  Closest match: WinSize: {best_params['win_size']}, kSize: {best_params['ksize']}, Alpha: {best_params['alpha']}, Corners: {best_corner_count}")
    else:
        print("  No suitable parameters found.")
    return best_win_size

def first_correct_scale(I):
    print(f"Task 2: Detecting corners across scales for image of shape: {I.shape}")
    I_blur = cv2.GaussianBlur(I, (5, 5), 0)
    psize = 6

    best_scale = -1
    best_corner_count = float('inf')
    best_params = {}

    for k in range(psize):
        J = I_blur.copy()
        for _ in range(k):
            J = cv2.pyrDown(J)
        gray = cv2.cvtColor(J, cv2.COLOR_BGR2GRAY)
        gray = np.float32(gray)

        for ksize_ in [3, 5, 7]:
            for alpha_ in [0.04, 0.06, 0.08]:
                dst = cv2.cornerHarris(gray, 4, ksize_, alpha_)
                dst = cv2.dilate(dst, None)
                thresh = 0.02 * dst.max()
                dst_thresh = (dst > thresh).astype(np.uint8) * 255

                num_labels, _, _, _ = cv2.connectedComponentsWithStats(dst_thresh)
                nc = num_labels - 1

                scale_factor = 2 ** k
                print(f"  Scale: {scale_factor}, kSize: {ksize_}, Alpha: {alpha_}, Corners: {nc}")

                if nc == NO_CORNERS:
                    print("  Found exact match!")
                    return scale_factor
                elif abs(nc - NO_CORNERS) < abs(best_corner_count - NO_CORNERS):
                    best_corner_count = nc
                    best_scale = scale_factor
                    best_params = {"ksize": ksize_, "alpha": alpha_}

    if best_scale != -1:
        print(f"  Closest match: Scale: {best_scale}, kSize: {best_params['ksize']}, Alpha: {best_params['alpha']}, Corners: {best_corner_count}")
    else:
        print("  No suitable parameters found.")
    return best_scale

# === Image Upload and Testing ===
from google.colab import files
uploaded = files.upload()

import io
from PIL import Image

# Automatically read the first uploaded image
image_name = next(iter(uploaded))
I = np.array(Image.open(io.BytesIO(uploaded[image_name])).convert('RGB'))
I = cv2.cvtColor(I, cv2.COLOR_RGB2BGR)  # Convert to OpenCV BGR format

cv2_imshow(I)
print("\nRunning WinSize Search...")
win_size = first_correct_winsize(I)
print("\nRunning Scale Search...")
scale = first_correct_scale(I)

print(f"\nFinal Results:\nBest WinSize: {win_size}\nBest Scale: {scale}")
```
