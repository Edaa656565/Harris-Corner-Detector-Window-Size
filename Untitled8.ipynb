```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

# Define the expected number of corners
NO_CORNERS = 78

# --- Task 1: first_correct_winsize function ---
def first_correct_winsize(I):
    """
    Finds the best window size and Harris parameters for detecting corners.
    """

    print(f"Task 1: Detecting corners for image of shape: {I.shape}")
    I_blur = cv2.GaussianBlur(I, (3, 3), 0)
    gray = cv2.cvtColor(I_blur, cv2.COLOR_BGR2GRAY)
    gray = np.float32(gray)

    best_win_size = -1
    best_corner_count = float('inf')
    best_params = {}

    for win_size_ in [2, 4, 8, 16]:  # Reduced range for speed
        for ksize_ in [3, 5]:  # Sobel kernel size
            for alpha_ in [0.04, 0.06]:  # Harris parameter
                dst = cv2.cornerHarris(gray, win_size_, ksize_, alpha_)
                _, dst_thresh = cv2.threshold(
                    dst, 0.01 * dst.max(), 255, cv2.THRESH_BINARY
                )
                dst_thresh = np.uint8(dst_thresh)

                num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(
                    dst_thresh
                )
                nc = num_labels - 1

                print(
                    f"  WinSize: {win_size_}, kSize: {ksize_}, Alpha: {alpha_},"
                    f" Corners: {nc}"
                )

                if nc == NO_CORNERS:
                    print("  Found exact match!")
                    return win_size_
                elif abs(nc - NO_CORNERS) < abs(best_corner_count - NO_CORNERS):
                    best_corner_count = nc
                    best_win_size = win_size_
                    best_params = {
                        "win_size": win_size_,
                        "ksize": ksize_,
                        "alpha": alpha_,
                    }

    if best_win_size != -1:
        print(
            "  No exact match. Returning closest match:"
            f" WinSize: {best_params['win_size']}, kSize:"
            f" {best_params['ksize']}, Alpha: {best_params['alpha']},"
            f" Corners: {best_corner_count}"
        )
    else:
        print("  No suitable parameters found.")
    return best_win_size

# --- Task 2: first_correct_scale function ---
def first_correct_scale(I):
    """
    Finds the best scale and Harris parameters for detecting corners across scales.
    """

    print(f"Task 2: Detecting corners across scales for image of shape: {I.shape}")
    I_blur = cv2.GaussianBlur(I, (3, 3), 0)
    psize = 6

    best_scale = -1
    best_corner_count = float('inf')
    best_params = {}

    for k in range(psize):
        J = I_blur.copy()
        for _ in range(k):  # Manual pyramid generation for parameter tuning
            J = cv2.pyrDown(J)
        gray = cv2.cvtColor(J, cv2.COLOR_BGR2GRAY)
        gray = np.float32(gray)

        for ksize_ in [3, 5]:
            for alpha_ in [0.04, 0.06, 0.08]:  # Expanded alpha range
                dst = cv2.cornerHarris(gray, 4, ksize_, alpha_)  # Fixed win_size
                _, dst_thresh = cv2.threshold(
                    dst, 0.01 * dst.max(), 255, cv2.THRESH_BINARY
                )
                dst_thresh = np.uint8(dst_thresh)

                num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(
                    dst_thresh
                )
                nc = num_labels - 1

                scale_factor = 2**k
                print(
                    f"  Scale: {scale_factor}, kSize: {ksize_}, Alpha: {alpha_},"
                    f" Corners: {nc}"
                )

                if nc == NO_CORNERS:
                    print("  Found exact match!")
                    return scale_factor
                elif abs(nc - NO_CORNERS) < abs(best_corner_count - NO_CORNERS):
                    best_corner_count = nc
                    best_scale = scale_factor
                    best_params = {"ksize": ksize_, "alpha": alpha_}

    if best_scale != -1:
        print(
            "  No exact match. Returning closest match:"
            f" Scale: {best_scale}, kSize: {best_params['ksize']},"
            f" Alpha: {best_params['alpha']}, Corners: {best_corner_count}"
        )
    else:
        print("  No suitable parameters found.")
    return best_scale

# --- Main execution block for Colab ---
print("Please upload '1.PNG' and '3.PNG' when prompted.")
uploaded = files.upload()

# Check if files were uploaded and load them
if "1.PNG" in uploaded and "3.PNG" in uploaded:
    I1 = cv2.imdecode(np.frombuffer(uploaded["1.PNG"], np.uint8), cv2.IMREAD_COLOR)
    I2 = cv2.imdecode(np.frombuffer(uploaded["3.PNG"], np.uint8), cv2.IMREAD_COLOR)

    if I1 is None or I2 is None:
        print(
            "Error: Could not load one or both images."
            " Please ensure they are valid image files."
        )
    else:
        # Resize images to have the same height before concatenation
        h1, w1, _ = I1.shape
        h2, w2, _ = I2.shape
        max_height = max(h1, h2)

        if h1 < max_height:
            I1_resized = cv2.resize(I1, (int(w1 * max_height / h1), max_height))
        else:
            I1_resized = I1

        if h2 < max_height:
            I2_resized = cv2.resize(I2, (int(w2 * max_height / h2), max_height))
        else:
            I2_resized = I2

        print("\n--- Running Task 1: first_correct_winsize ---")
        s1 = first_correct_winsize(I1_resized)
        s2 = first_correct_winsize(I2_resized)

        if s1 != -1 and s2 != -1:
            J_winsize = np.concatenate((I1_resized, I2_resized), axis=1)
            txt_winsize = ""
            if s1 < s2:
                txt_winsize = (
                    f"Logo 1 is {s2 / s1:.2f} times smaller than logo 2"
                    f" (WinSize: {s1} vs {s2})"
                )
            elif s1 > s2:
                txt_winsize = (
                    f"Logo 1 is {s1 / s2:.2f} times larger than logo 2"
                    f" (WinSize: {s1} vs {s2})"
                )
            else:
                txt_winsize = (
                    "Logo 1 is about the same size as logo 2" f" (WinSize: {s1} vs {s2})"
                )
            cv2.putText(
                J_winsize, txt_winsize, (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2
            )
            print(f"\nResult for Task 1: {txt_winsize}")
            cv2_imshow(J_winsize)
        else:
            print(
                "\nTask 1: Could not determine logo sizes due to"
                " inability to find suitable window sizes."
            )
            if 'J_winsize' in locals(): # Ensure J_winsize was created
                cv2_imshow(J_winsize) # Display even if text wasn't added

        print("\n--- Running Task 2: first_correct_scale ---")
        sc1 = first_correct_scale(I1_resized.copy())
        sc2 = first_correct_scale(I2_resized.copy())

        if sc1 != -1 and sc2 != -1:
            J_scale = np.concatenate((I1_resized, I2_resized), axis=1)
            txt_scale = ""
            if sc1 < sc2:
                txt_scale = (
                    f"Logo 1 is {sc2 / sc1:.2f} times smaller than logo 2"
                    f" (Scale: {sc1} vs {sc2})"
                )
            elif sc1 > sc2:
                txt_scale = (
                    f"Logo 1 is {sc1 / sc2:.2f} times larger than logo 2"
                    f" (Scale: {sc1} vs {sc2})"
                )
            else:
                txt_scale = (
                    "Logo 1 is about the same size as logo 2" f" (Scale: {sc1} vs {sc2})"
                )
            cv2.putText(
                J_scale, txt_scale, (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2
            )
            print(f"\nResult for Task 2: {txt_scale}")
            cv2_imshow(J_scale)
        else:
            print(
                "\nTask 2: Could not determine logo sizes due to"
                " inability to find suitable scales."
            )
            if 'J_scale' in locals(): # Ensure J_scale was created
                cv2_imshow(J_scale) # Display even if text wasn't added

else:
    print("Error: '1.PNG' or '3.PNG' not found. Please upload both files.")
```
